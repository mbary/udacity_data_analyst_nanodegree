```python
plt.plot(df_city.year,df_city.avg_temp)
plt.plot(df_global.year, df_global.avg_temp)
```

```python
df_city["5y_mean"] = df_city.avg_temp.rolling(window=5).mean()
df_city["10y_mean"] = df_city.avg_temp.rolling(window=10).mean()
df_city["15y_mean"] = df_city.avg_temp.rolling(window=15).mean()
```

```python
df_global["5y_mean"] = df_global.avg_temp.rolling(window=5).mean()
df_global["10y_mean"] = df_global.avg_temp.rolling(window=10).mean()
df_global["15y_mean"] = df_global.avg_temp.rolling(window=15).mean()
```

```python
plt.figure(figsize=(16,8))
plt.title("Warsaw vs Global avg. temp. 5-year rolling avg", weight = "bold", fontsize = 15)

plt.plot(df_city.year, df_city["5y_mean"], label = "Warsaw")
plt.plot(df_global.year , df_global["5y_mean"] ,label = "Global")

plt.xlabel("Year", weight = "bold", fontsize = 13)
plt.ylabel("Average Temperature", weight = "bold", fontsize = 13)
plt.legend(fontsize=15)
# plt.savefig("5y_avg.png");
```

```python
plt.figure(figsize = (16,8))
plt.title("Warsaw vs Global avg. temp. 10-year rolling avg", weight="bold", fontsize=15)

plt.plot(df_city.year, df_city["10y_mean"], label = "Warsaw")
plt.plot(df_global.year, df_global["10y_mean"], label = "Global")

plt.xlabel("Year", weight = "bold", fontsize=13)
plt.ylabel("Average Temperature", weight = "bold", fontsize = 13)
plt.legend(fontsize=15)
# plt.savefig("10y_avg.png");
```

```python
plt.figure(figsize = (16,8))
plt.title("Warsaw vs Global avg. temp. 15-year rolling avg", weight = "bold", fontsize = 15)

plt.plot(df_city.year, df_city["15y_mean"], label = "Warsaw")
plt.plot(df_global.year, df_global["15y_mean"], label = "Global")

plt.xlabel("Year", weight = "bold", fontsize=13)
plt.ylabel("Average Temperature", weight = "bold", fontsize = 13)
plt.legend(fontsize=15)
# plt.savefig("15y_avg.png");
```
